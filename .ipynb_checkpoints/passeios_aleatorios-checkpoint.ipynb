Inicialmente vamos considerar
um passeio aleatório simples que comece em 0, com passos de 1 e
–1 ocorrendo com a mesma probabilidade.

```python
import numpy as np
import matplotlib.pyplot as plt
total = 1000
possibilidades = np.random.randint(0,2,size=total)
passos = np.where(possibilidades>0,1,-1)
passeio = passos.cumsum() # soma cumulativa
plt.plot(passeio[:100])
```
